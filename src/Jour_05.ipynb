# Visualisation avec scatter plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le jeu de donn√©es Iris
iris = sns.load_dataset('iris')
iris.head()
```

```python
# 1. Scatter plot simple
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter plot: Sepal Length vs Sepal Width')
plt.show()
```

```python
# 2. Pairplot
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# 3. Scatter plot avec r√©gression
plt.figure(figsize=(10, 6))
sns.lmplot(data=iris, x='petal_length', y='petal_width', hue='species', height=6)
plt.title('Scatter plot avec r√©gression: Petal Length vs Petal Width')
plt.show()
```

```python
# 4. Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Box plot: Sepal Length par esp√®ce')
plt.show()
```

```python
# 5. Violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=iris, x='species', y='petal_width')
plt.title('Violin plot: Petal Width par esp√®ce')
plt.show()
```
